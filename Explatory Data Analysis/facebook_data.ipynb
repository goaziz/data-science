```python
plt.figure(figsize=(12, 7))
sns.countplot(x='gender', data=df)
```

### The graph above shows males are likely more to use FB than females

```python
plt.figure(figsize=(20, 10))
sns.countplot(x='age', data=df)
plt.tight_layout()
```

### It seems that as soon as people get old they use facebook very little 

```python
print('Mean of likes receved on mobile is', df['mobile_likes_received'].mean())
print('Mean of likes receved on www is', df['www_likes_received'].mean())
```

```python
plt.figure(figsize=(12, 7))
plt.xlabel('Age')
plt.ylabel('Likes received')
sns.lineplot(x='age', y='likes_received', data=df)
```

### Between age group of 13 and 20 the more likes were received. In addition, from 22 to 65 the rend did not change drammatically, this age group of people received approximately 100 likes but it fluctuated seriously between 73 and 113

```python
plt.figure(figsize=(12, 7))
plt.xlabel('Age')
plt.ylabel('Likes received')
sns.lineplot(x='age', y='likes_received', data=df, hue='gender')
```

### Females whose age is between 13 and 70 got more likes than males 

```python
plt.figure(figsize=(12, 7))
plt.xlabel('Friend Count')
plt.ylabel('Birth Year')
plt.title('People friend count on Facebook')
sns.lineplot(data=df, x="dob_year", y="friend_count", hue='gender')
plt.show()
```

### Women had slightly more friends on facebook than males

```python
df['likes_per_day'] = df['likes_received'] / df['tenure'].where(df['tenure'] > 0)
```
